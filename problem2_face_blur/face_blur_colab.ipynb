### Inline Video Player
Colab doesn't natively play H.264 MP4 videos easily without converting them to a standard format that browser HTML5 players support. Let's define a helper to display the video.

```python
# 4. Display the output video inline
from IPython.display import HTML
from base64 import b64encode

# We compress/re-encode the output for HTML5 web compatibility in Colab
!ffmpeg -i sample_output.mp4 -vcodec libx264 -f mp4 web_output.mp4 -y -loglevel quiet

mp4 = open('web_output.mp4','rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML(f"""
<video width=600 controls>
      <source src="{data_url}" type="video/mp4">
</video>
""")
```
