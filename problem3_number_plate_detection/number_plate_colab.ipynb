### View JSON Results

```python
# Print sample detections from JSON log
import json
with open('sample_output.json') as f:
    data = json.load(f)

print("Sample Detections:")
frames_with_detections = {k: v for k, v in data.get('frames', {}).items() if v}
for i, (frame, dets) in enumerate(list(frames_with_detections.items())[:10]):
    print(f"  Frame {frame}: {dets}")
```

### Inline Video Player
We re-encode the output video with x264 so it can be streamed directly inside the browser using HTML5.

```python
# 4. Display the output video inline
from IPython.display import HTML
from base64 import b64encode

!ffmpeg -i sample_output.mp4 -vcodec libx264 -f mp4 web_output.mp4 -y -loglevel quiet

mp4 = open('web_output.mp4','rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML(f"""
<video width=600 controls>
      <source src="{data_url}" type="video/mp4">
</video>
""")
```
